##**Exp2**
RGB to grayscale

In this exercise you will load an image from scikit-image module data and make it grayscale, then compare both of them in the output. We have preloaded a function show_image(image, title='Image') that displays the image using Matplotlib. You can check more about its parameters using ?show_image() or help(show_image) in the console.

```python
from skimage import data,color
import matplotlib.pyplot as plt
# Load the rocket image
rocket_img=data.rocket()
#conver to RGB to gray
r_gray=color.rgb2gray(rocket_img)
plt.imshow(rocket_img);
plt.axis("off")
plt.title("Original image")
plt.show()
plt.imshow(r_gray,cmap='gray')
plt.axis("off")
plt.title("Gray converted")
plt.show()

```

### 4.1: Global Thresholding

```python
chess=plt.imread('/content/drive/MyDrive/MY_Learnings/Python4Imgprocessing_Datacamp/chess.jpg')
showimage(chess,'Original Image','jet')
chess_gray=color.rgb2gray(chess)
showimage(chess_gray,'Grayscale Image','gray')
from skimage.filters import threshold_otsu
Th=threshold_otsu(chess_gray)
binary=chess_gray>Th
showimage(binary,'Binarized','gray')
```

### 4.2 Local Thresholding : This is applicable only if the image is having uneven background. Else, the result will be a complete balck image.

```python
from skimage.filters import threshold_local
page=plt.imread('/content/drive/MyDrive/MY_Learnings/Python4Imgprocessing_Datacamp/page.png')
page=color.rgb2gray(page)
showimage(page,'original','gray')
block=35;
threslocal=threshold_local(page,block,offset=1)
binary_local1=page>threslocal
#showimage(binary_local1,'Localbinarization','gray')
plt.imshow(binary_local1)
```

4.3 All Threshold

```python
fruits_image=plt.imread('/content/drive/MyDrive/MY_Learnings/Python4Imgprocessing_Datacamp/fruits-2.jpg')
# Import the try all function
from skimage.filters import try_all_threshold

# Import the rgb to gray convertor function 
from skimage.color import rgb2gray

# Turn the fruits_image to grayscale
grayscale = rgb2gray(fruits_image)

# Use the try all method on the resulting grayscale image
fig, ax = try_all_threshold(grayscale, verbose=False)

# Show the resulting plots
plt.show()
```

**Apply thresholding:**
In this exercise, you will decide what type of thresholding is best used to binarize an image of knitting and craft tools. In doing so, you will be able to see the shapes of the objects, from paper hearts to scissors more clearly.

```python
tools_image=plt.imread('/content/drive/MyDrive/MY_Learnings/Python4Imgprocessing_Datacamp/shapes52.jpg')
# Import threshold and gray convertor functions
from skimage.filters import threshold_otsu
from skimage.color import rgb2gray

# Turn the image grayscale
gray_tools_image = rgb2gray(tools_image)

# Obtain the optimal thresh
thresh = threshold_otsu(gray_tools_image)

# Obtain the binary image by applying thresholding
binary_image = gray_tools_image > thresh

# Show the resulting binary image
showimage(binary_image, 'Binarized image','gray')
```
